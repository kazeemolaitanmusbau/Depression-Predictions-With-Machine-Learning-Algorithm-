# Data Explaration Analysis

```python
for i in df1.select_dtypes("object").columns:
    df1[i].value_counts().plot(kind="bar")
    plt.title(f"count of {i}")
    plt.show()
```

```python
for i in df1.select_dtypes(["int", "float"]).columns:
    df1[i].value_counts().plot(kind="hist")
    plt.title(f"distributin of  of {i}")
    plt.show()
```
